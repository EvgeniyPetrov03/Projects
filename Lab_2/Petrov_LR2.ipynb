```python
import warnings
warnings.filterwarnings('ignore')
data = pd.read_csv('trip.csv')
data = data.sort_values(by='birthyear')
data = data[(data['birthyear'] >= 1931) & (data['birthyear']<=1999)]
groupby_birthyear = data.groupby('birthyear').size()
groupby_birthyear.plot.bar(title = 'Distribution of birth years', figsize = (15,4))
plt.show()
```

```python
import datetime
data['StartTime1'] = data['starttime'].apply(lambda x: datetime.datetime.strptime(x, "%m/%d/%Y %H:%M") )
data['year'] = data['StartTime1'].apply(lambda x: x.year )
data['month'] = data['StartTime1'].apply(lambda x: x.month )
data['day'] = data['StartTime1'].apply(lambda x: x.day )
data['hour'] = data['StartTime1'].apply(lambda x: x.hour )
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
diabetes = datasets.load_diabetes()
```

```python
X_train = diabetes.data[:-20]
X_test = diabetes.data[-20:]
y_train = diabetes.target[:-20]
y_test = diabetes.target[-20:]
```

```python
regr = LinearRegression()
```

```python
regr.fit(X_train, y_train)
```

```python
y_pred = regr.predict(X_test)
print ('R2 score: %f' %r2_score(y_test, y_pred))
print ('Intercept: %f' %regr.intercept_)
print ('Coefficients: %s' %str(regr.coef_))
```

```python
plt.scatter(X_test[:,3], y_test, color='black')
plt.plot(X_test[:,3], y_pred, color='red', linewidth=3)
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
from sklearn.svm import SVR
regr = SVR(kernel='linear')
```

```python
regr.fit(X_train, y_train)
```

```python
y_pred = regr.predict(X_test)
print ('R2 score: %f'%r2_score(y_test, y_pred))
print ('Intercept: %f'%regr.intercept_)
#print ('Coefficients: %s'%str(regr.coef_))
plt.scatter(X_test[:,3], y_test, color='black')
plt.plot(X_test[:,3], y_pred, color='blue', linewidth=3)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.ensemble import RandomForestClassifier
```

```python
y_train = np.where(y_train > 100, 1, -1)
y_test = np.where(y_test > 100, 1, -1)
```

```python
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, y_train)
```

```python
accuracy = rf.score(X_test, y_test)
print ('Accuracy: %f'%accuracy)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv('train.csv')
MedianAge = data.groupby('Age').count()['PassengerId']
data.Age[data.Age.isnull()] = MedianAge[MedianAge == MedianAge.median()].index[0]
data['Sex'].replace(['male','female'],[1,0],inplace=True)
```

```python
MedianS = data.groupby('Survived').count()['PassengerId']
data.Survived[data.Survived.isnull()] = MedianAge[MedianAge == MedianAge.median()].index[0]

MedianP = data.groupby('Pclass').count()['PassengerId']
data.Pclass[data.Pclass.isnull()] = MedianAge[MedianAge == MedianAge.median()].index[0]

MedianSex = data.groupby('Sex').count()['PassengerId']
data.Sex[data.Sex.isnull()] = MedianAge[MedianAge == MedianAge.median()].index[0]

MedianPar = data.groupby('Age').count()['PassengerId']
data.Age[data.Age.isnull()] = MedianAge[MedianAge == MedianAge.median()].index[0]

MedianPar = data.groupby('Parch').count()['PassengerId']
data.Parch[data.Parch.isnull()] = MedianAge[MedianAge == MedianAge.median()].index[0]

MedianSib = data.groupby('SibSp').count()['PassengerId']
data.SibSp[data.SibSp.isnull()] = MedianAge[MedianAge == MedianAge.median()].index[0]

data = data.drop(['PassengerId','Name','Ticket','Fare', 'Cabin', 'Embarked'],axis=1)
```

```python
X_train = data[:-20]
X_test = data[-20:]
y_train = data.Survived[:-20]
y_test = data.Survived[-20:]
regr = LinearRegression()
```

```python
regr.fit(X_train, y_train)
```

```python
y_pred = regr.predict(X_test)
print ('R2 score: %f' %r2_score(y_test, y_pred))
print ('Intercept: %f' %regr.intercept_)
print ('Coefficients: %s' %str(regr.coef_))
```
